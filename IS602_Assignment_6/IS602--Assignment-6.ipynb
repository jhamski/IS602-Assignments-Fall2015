```python
import numpy as np
import timeit
import random
```

In Homework 1 we used list of about six integers long. For this assignment I created a much larger sample list of 1e4 numbers randomly sampled from 0 to 1e7. This large integer data should demonstrate computation time more effectively. 

```python
L_list = random.sample(xrange(10000000), 10000)
L_np_array = np.array(L_list)
```

```python
setup = '''
import copy

def sortwithoutloops(input):
    input.sort()
    return input
'''
```

```python
n = 100
```

```python
t = timeit.Timer("x=copy.copy(L_list);sortwithoutloops(x)",setup=setup)
print 'sortwithoutloops: ',t.timeit(n)
```
