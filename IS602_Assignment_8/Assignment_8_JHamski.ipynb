```python
%pylab inline

import numpy as np
#import pylab
import matplotlib.pylab
import mahotas as mh
from scipy import ndimage
```

## Image 1: Circles

```python
circles = mh.imread('circles.png')
pylab.imshow(circles)
pylab.show()
```

```python
T = mh.otsu(circles)
circles_th = (circles > T)
circles_sm = mh.gaussian_filter(circles_th, 20)
circles_regmax = mh.regmax(circles_sm)
labels, nr_obj = mh.label(circles_regmax)
pylab.imshow(mh.overlay(circles, circles_regmax))
pylab.show()
```

```python
km,circles_cnt = mh.label(circles_regmax)
print circles_cnt
```

```python
centers = mh.center_of_mass(circles, labels)[1:]
print centers
```
