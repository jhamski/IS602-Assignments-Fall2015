```python
%pylab inline

import numpy as np
#import pylab
import matplotlib.pylab
import mahotas as mh
from scipy import ndimage
```

```python
circles = mh.imread('circles.png')
```

```python
pylab.imshow(circles)
pylab.show()
```

```python
T = mh.otsu(circles)
circles_th = (circles > T)
circles_sm = mh.gaussian_filter(circles_th, 20)
circles_regmax = mh.regmax(circles_sm)
labeled, nr_objects = mh.label(circles_regmax)
pylab.imshow(mh.overlay(circles, circles_regmax))
pylab.show()
```
