```python
%pylab inline

import numpy as np
#import pylab
import matplotlib.pylab
import mahotas as mh
from scipy import ndimage
```

```python
circle = mh.imread('circles.png')
```

```python
pylab.imshow(circle)
pylab.show()
```

```python
T = mh.thresholding.rc(circle)
pylab.imshow(circle > T)
pylab.show()
```

```python
gauss_denoised = ndimage.gaussian_filter(circle, 1)
T = mh.thresholding.otsu(gauss_denoised)
pylab.imshow(gauss_denoised>T)
pylab.show()
```

```python
labeled,nr_objects = mh.label(gauss_denoised)
print nr_objects
pylab.imshow(labeled)
pylab.jet()
pylab.show()
```
